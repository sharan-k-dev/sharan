```python
from transformers import pipeline

sentiment = pipeline("sentiment-analysis")
sentiment("i am happy about this course")
```

**Text generation pipeline:**


```python
classifier = pipeline("text-generation",model = 'distilgpt2')
classifier('i am sharan, an aspiring AI Engineer')
```

**zero-shot-classification:**

```python
classifier = pipeline("zero-shot-classification")
classifier("This course is awesome",candidate_labels = ['education','politics'])
```

**Fill mask**

```python
classifier = pipeline('fill-mask')
classifier('i am learning transformers on the <mask> provided by huggibg face',top_k =2)
```

Summarization:

```python
classifier = pipeline('summarization', model = 'distilgpt2')
classifier('Artificial Intelligence (AI) is a branch of computer science that focuses on creating machines capable of mimicking human intelligence. It involves developing algorithms that enable computers to perform tasks such as problem-solving, learning, decision-making, and natural language processing. AI is categorized into narrow AI, which is designed for specific tasks like voice assistants and recommendation systems, and general AI, which aims to replicate human cognitive abilities across various domains. Machine learning and deep learning, subsets of AI, allow systems to improve performance over time by analyzing data patterns. AI is transforming industries such as healthcare, finance, education, and transportation, enhancing efficiency and automation. However, ethical concerns, including data privacy, bias, and job displacement, remain critical challenges. As AI continues to evolve, balancing innovation with ethical considerations is essential for its responsible development and deployment.')
```
